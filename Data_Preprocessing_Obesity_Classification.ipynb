## Outlier Detection

```python
df.head()
```

```python
sns.boxplot(df.Age)
```

```python
q1 = df.Age.quantile(0.25)   # Q1

q3 = df.Age.quantile(0.75)   #Q3
```

```python
q1
```

```python
q3
```

```python
# (Inter Quartile Range)
IQR = q3-q1
IQR
```

```python
Upper_limit = q3+1.5*IQR
Upper_limit
```

```python
lower_limit = q1 - 1.5*IQR
lower_limit
```

```python
df.Age.median()
```

```python
df['Age'] = np.where(df['Age']>Upper_limit, 42.5,df['Age'])
```

```python
sns.boxplot(df.Age)
```
